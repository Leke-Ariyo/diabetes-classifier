### Trimming the dataset

```python
# Due to the fact that the dataset is large and highly imbalance, it will be trimmed.

# no diabetes
df = df.sample(frac = 1)
no_d = df[df['Diabetes_012'] == 0]
sel_no_d = no_d[:7000]

# prediabetes
df = df.sample(frac = 1)
pre_d = df[df['Diabetes_012'] == 1]
sel_pre_d = pre_d[:5000]

# diabetes
df = df.sample(frac = 1)
d = df[df['Diabetes_012'] == 2]
sel_d = d[:4500]

# concatenating the dataframes
frames = [sel_no_d, sel_pre_d, sel_d]
df_ = pd.concat(frames).reset_index(drop=True)
df_.head(10)
```

```python
sel_feat = ['HighChol', 'HighBP', 'Age', 'HeartDiseaseorAttack', 'DiffWalk', 'GenHlth', 'PhysHlth', 'Stroke', 'AnyHealthcare']
```

```python
# Scaler function for scaling the dataset
from sklearn.preprocessing import MinMaxScaler
def scaler(feature):
    scale = MinMaxScaler()
    scaled = scale.fit_transform(feature)
    return scaled
```

```python
# getting the target variable and then dropping it from the dataset
target = df_['Diabetes_012']
df_1 = df_[sel_feat]
```

```python
# Scaling all the features
for column in df_1.columns:
    df_1[column] = scaler(df_1[[column]])
```

### Unsupervised Learning - K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X = df_1.iloc[:, :].values
```

### Elbow method

```python
# finding the best clusters
err_rng = range(2, 16)
error = []
for i in range(2, 16):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    sse = kmeans.inertia_
    error.append(sse)
```

```python
plt.figure(figsize=(15, 8))
plt.xlabel('K', family='Arial', fontsize = 14, color='black')
plt.ylabel('sum of square error',family='Arial', fontsize = 14, color='black')
plt.xticks(fontsize=12, color = 'black')
plt.yticks(fontsize=12, color = 'black')
plt.grid(which = 'both', color='black', axis = 'x', alpha=0.5)
plt.plot(err_rng, error)
```

### Silhouette Score

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(X) for k in range(2, 16)]

silhouette_scores = [silhouette_score(X, model.labels_) for model in kmeans_per_k[1:]]
```

```python
silhouette_scores
```

```python
plt.figure(figsize=(15, 8))
plt.xlabel('K', family='Arial', fontsize = 14, color='black')
plt.ylabel('Silhouette Score',family='Arial', fontsize = 14, color='black')
plt.grid(which = 'both', color='black', axis = 'x', alpha=0.5)
plt.plot(range(2, 15), silhouette_scores, color='blue', linewidth=3)
```

### Supervised Learning 

```python
# Supervised Learning Approach.

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df_1, test_size=0.2, random_state=42)
train_label, test_label = train_test_split(target, test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier()
rf.fit(train, train_label)
pred = rf.predict(test)
acc = accuracy_score(test_label, pred)
print(acc)
print(classification_report(test_label, pred))
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(test_label, pred)
print(cm)
sns.heatmap(cm, annot=True)
```
